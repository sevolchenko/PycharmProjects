```python
from Oscillation import *

global first
first = MechanicOscillation(x0=0.05, m=1, k=5)
print('first oscillogramm parameters:\n{0}\n'.format(str(first)))
global second
second = ElectricOscillation(q0=0.05, capacitance=0.005, inductance=1)
print('second oscillogramm parameters:\n{0}\n'.format(str(second)))
global third
third = Oscillation(a0=0.05, damping_fact=0.2, freq=0.5)
print('third oscillogramm parameters:\n{0}\n'.format(str(third)))
```

```python
import numpy as np

time = np.arange(0, 20, 0.01)

global first_values
first_values = [first.x(t1) for t1 in time]
global second_values
second_values = [second.x(t1) for t1 in time]
global third_values
third_values = [third.x(t1) for t1 in time]
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.plot(time, first_values, 'b', label="first")
ax.plot(time, second_values, 'r', label="second")
ax.plot(time, third_values, 'g', label="third")

ax.set(xlabel='t', ylabel='x(t)', title='x(t) = a0 * e ** (-b * t) * sin(w * t + alpha)')
ax.legend()
ax.grid()
plt.show()
```
